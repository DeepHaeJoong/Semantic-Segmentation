### 1.1 FCN with Resnet-101 backbone

```python
from torchvision import models

fcn = models.segmentation.fcn_resnet101(pretrained=True).eval()
```

### 1.2 Load the Image

```python
from PIL import Image
import matplotlib.pyplot as plt
import torch
#import wget

#!wget -nv https://static.independent.co.uk/s3fs-public/thumbnails/image/2018/04/10/19/pinyon-jay-bird.jpg -O bird.png
img = Image.open('./bird.png')
plt.imshow(img); plt.show()
```

### 1.3 Pre-process the image

- mean = [0.485, 0.456, 0.406]
- std = [0.229, 0.224, 0.225]

```python
# Apply the transformations needed
import torchvision.transforms as T

trf = T.Compose([T.Resize(256),
                 T.CenterCrop(224),
                 T.ToTensor(), 
                 T.Normalize(mean = [0.485, 0.456, 0.406], 
                             std = [0.229, 0.224, 0.225])])

inp = trf(img).unsqueeze(0)
```

### 1.4. Forward pass through the network

```python
# Pass the input through the net
out = fcn(inp)['out']
print (out.shape)
```

```python
import numpy as np
om = torch.argmax(out.squeeze(), dim=0).detach().cpu().numpy()
print (om.shape)
```

```python
print (np.unique(om))
```

### 1.5 Decode Output

```python
# Define the helper function

def decode_segmap(image, nc=21):
  
    label_colors = np.array([(0, 0, 0),  # 0=background
               # 1 = aeroplane, 2 = bicycle, 3 = bird, 4 = boat, 5 = bottle
               (128, 0, 0), (0, 128, 0), (128, 128, 0), (0, 0, 128), (128, 0, 128),
               # 6 = bus, 7 = car, 8 = cat, 9 = chair, 10 = cow
               (0, 128, 128), (128, 128, 128), (64, 0, 0), (192, 0, 0), (64, 128, 0),
               # 11 = dining table, 12 = dog, 13 = horse, 14 = motorbike, 15 = person
               (192, 128, 0), (64, 0, 128), (192, 0, 128), (64, 128, 128), (192, 128, 128),
               # 16 = potted plant, 17 = sheep, 18 = sofa, 19 = train, 20 = tv/monitor
               (0, 64, 0), (128, 64, 0), (0, 192, 0), (128, 192, 0), (0, 64, 128)])

    r = np.zeros_like(image).astype(np.uint8)
    g = np.zeros_like(image).astype(np.uint8)
    b = np.zeros_like(image).astype(np.uint8)

    for l in range(0, nc):
        idx = image == l
        r[idx] = label_colors[l, 0]
        g[idx] = label_colors[l, 1]
        b[idx] = label_colors[l, 2]

    rgb = np.stack([r, g, b], axis=2)
    
    return rgb
```

```python
rgb = decode_segmap(om)
plt.imshow(rgb); plt.show()
```

### 1.6 Final Result

```python
def segment(net, path, show_orig=False):
    img = Image.open(path)
    
    plt.imshow(img); plt.axis('off'); plt.show()
    
    # Comment the Resize and CenterCrop for better inference results
    trf = T.Compose([T.Resize(256), 
                   T.CenterCrop(224), 
                   T.ToTensor(), 
                   T.Normalize(mean = [0.485, 0.456, 0.406], 
                               std = [0.229, 0.224, 0.225])])
    
    inp = trf(img).unsqueeze(0)
    out = net(inp)['out']
    om = torch.argmax(out.squeeze(), dim=0).detach().cpu().numpy()
    rgb = decode_segmap(om)
    plt.imshow(rgb); plt.axis('off'); plt.show()
```

```python
#!wget -nv https://images.pexels.com/photos/1996333/pexels-photo-1996333.jpeg -O horse.png
segment(fcn, './horse.png')
```
